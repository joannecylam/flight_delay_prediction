```python
import pandas as pd
from matplotlib import pyplot as plt
from jupyterthemes import jtplot
jtplot.style(theme='grade3')
```

```python
train_data = pd.read_csv('flight_delays_data.csv')
train_data = train_data[['flight_id', 'is_claim']]
```

```python
result_df = pd.read_csv('results_1129_100_true.csv', header=None, names=['flight_id', 'predictions'])
```

```python
result_df['predictions'] = result_df['predictions'].astype(float)
```

```python
result_df['predicted_claim'] = [float(amt)*800 for amt in result_df['predictions'].values]
```

```python
join_df = pd.merge(result_df, train_data, on='flight_id')
```

```python
sub_df = join_df.head(10000).sort_values('flight_id')
plt.scatter(sub_df['is_claim'].values, sub_df['predicted_claim'].values)
plt.xlabel("True Value")
plt.ylabel("Predicted Value")
plt.title("Random Forest")
plt.show()
```
